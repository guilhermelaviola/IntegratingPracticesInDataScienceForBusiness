# **Descriptive Data Analysis**
Descriptive data analysis is the transformation of raw data into descriptive information, whether through text or graphical visualizations. This process is crucial for business, as it allows managers and executives to visualize data clearly and efficiently. For example, through data visualization, it's possible to identify where the highest sales occur, which regions have the highest demand, and which salespeople are excelling.

## **Performing Descriptive Analysis in Python**
In situations in which we have to deal with larger and more complex databases, Excel can become limited. Therefore, we'll use Python, a powerful programming language that offers several libraries specifically for data analysis.

```python
# Importing all the necessary libraries:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating a DataFrame with sales data:
date = pd.date_range(start = '2025-01-01', periods=12, freq='M')
sales = [156, 118, 235, 96, 113, 345, 286, 271, 254, 311, 276, 235]

df = pd.DataFrame({'Date': date, 'Sales': sales})
```

```python
# Setting the chart:
sns.set(style='whitegrid')

# Creating the line chart to vusualize the sales trend:
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Sales', data=df, marker='o')

# Adding titles and labels:
plt.title('Sale trends within time:')
plt.xlabel('Date')
plt.ylabel('Sales')

# Displaying the chart:
plt.show()
```

## **Trend and Pattern Analysis**
Trend analysis allows us to predict future behavior based on historical data. This is crucial for planning growth strategies and identifying areas for improvement. Using tools like Seaborn and Matplotlib, we can visualize these trends clearly and efficiently.

```python
# Creating a dataset with employee fictional data:
data = {
    'Name': ['Anne', 'Kevin', 'Charles', 'Danielle', 'Edward'],
    'Score': [88, 92, 75, 85, 79],
    'Absenses': [2, 0, 4, 1, 3]
}

# Creating a DataFrame with the data above:
df = pd.DataFrame(data)
```

```python
# Displaying basic information of the DataFrame:
print(df.info())

# Generating descriptive statistics:
print(df.describe())
```

```python
# Configuring the stype of the chart:
sns.set(style="whitegrid")

# Creating an histogram of the score distribution:
plt.figure(figsize=(8, 6))
sns.histplot(df['Score'], bins=5, kde=True)

# Adding title and labels:
plt.title('Employee Score Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')

# Displaying the chart:
plt.show()
```

```python
# Creating a scatterplot for scores and absenses:
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Absenses', y='Score', data=df)

# Adding titles and labels:
plt.title('Relationship between Score and Absenses')
plt.xlabel('Absenses')
plt.ylabel('Score')

# Displaying the chart:
plt.show()
```

## **Practical Application of Descriptive Data Analysis**
We'll use Python, along with the Seaborn, Matplotlib, and Pandas libraries, to perform a comprehensive analysis. While it may seem complicated, descriptive analysis with Python is quite coherent and meets the needs of different types of data, especially when dealing with large volumes.

```python
# Creating a dataset with fictional data:
data = pd.date_range(start='2024-01-01', periods=12, freq='M')
sales = [1500, 1800, 1700, 1900, 2100, 2200, 2000, 2300, 2500, 2400, 2600, 2700]
expenses = [1200, 1300, 1100, 1400, 1500, 1600, 1550, 1650, 1700, 1800, 1900, 2000]

# Creating a DataFrame with the data above:
df = pd.DataFrame({'Data': data, 'Sales': sales, 'Expenses': expenses})
```

```python
# Configuring the style of the chart:
sns.set(style="whitegrid")

# Creating a lineplot to display sales and expenses over time
plt.figure(figsize=(10, 6))

# Sales chart
sns.lineplot(x='Data', y='Sales', data=df, label='Sales', marker='o')

# Expenses chart:
sns.lineplot(x='Data', y='Expenses', data=df, label='Expenses', marker='o')

# Adding title and labels:
plt.title('Sales and Expenses over time')
plt.xlabel('Data')
plt.ylabel('Expenses (in USD)')
plt.legend()

# Displaying the chart:
plt.show()
```
